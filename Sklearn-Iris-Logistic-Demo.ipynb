```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
```

```python
iris_data = load_iris() # load dataset to get data + metadata

print(iris_data.feature_names) # X columns
print(iris_data.target_names) # y result
```

```python
# destructure the dataset
X_iris = iris_data.data
y_iris = iris_data.target

# for plots
x_col = 0
y_col = 1

# Visualizations to see clusters
plt.scatter(X_iris[:,x_col], X_iris[:,y_col], c=y_iris)
plt.xlabel(iris_data.feature_names[x_col])
plt.ylabel(iris_data.feature_names[y_col])
plt.show()

x_col = 2
y_col = 3
plt.scatter(X_iris[:,x_col], X_iris[:,y_col], c=y_iris)
plt.xlabel(iris_data.feature_names[x_col])
plt.ylabel(iris_data.feature_names[y_col])
plt.show()

x_col = 1
y_col = 3
plt.scatter(X_iris[:,x_col], X_iris[:,y_col], c=y_iris)
plt.xlabel(iris_data.feature_names[x_col])
plt.ylabel(iris_data.feature_names[y_col])
plt.show()

```

```python
# build classifier
model = LogisticRegression(random_state=1, max_iter = 125).fit(X_iris, y_iris) 
    # increase max iterations to reach convergence
```

```python
# choose a subset of random data rows and predict
print(model.predict(X_iris[60:100, :]))
print(model.predict_proba(X_iris[60:100,:]))
```

```python
model.score(X_iris, y_iris)
```

```python
# manually calculate accuracy
preds = model.predict(X_iris[0:y_iris.size, :])
mistakes = 0
for i in range(y_iris.size):
    if(preds[i] != y_iris[i]):
        mistakes += 1
print("Mistakes: " + str(mistakes))
print("Accuracy: " + str(1-mistakes/y_iris.size)) #results in same output as score function
```
